```python
def avgRating(dataFrame):
    dataFrame['avgRating'] = (np.median(dataFrame['OnOffRtg']) + np.median(dataFrame['OnDefRtg'])) / 2

def teamRatings(dataFrame):
    tempData = dataFrame.copy()
    teamMinutes = tempData.groupby('TeamAbbreviation')['Minutes'].sum()
    for team in teamMinutes.index:
        tempData.loc[tempData['TeamAbbreviation'] == team, 'percentageOfTeamMinutes'] = tempData['Minutes'] / teamMinutes[team]

    tempData['tempOff'] = tempData['OnOffRtg'] * tempData['percentageOfTeamMinutes']
    teamOff = tempData.groupby('TeamAbbreviation')['tempOff'].sum()
    for team in teamOff.index:
        dataFrame.loc[dataFrame['TeamAbbreviation'] == team, 'teamOffRating'] = teamOff[team]

    tempData['tempDef'] = tempData['OnDefRtg'] * tempData['percentageOfTeamMinutes']
    teamDef = tempData.groupby('TeamAbbreviation')['tempDef'].sum()
    for team in teamDef.index:
        dataFrame.loc[dataFrame['TeamAbbreviation'] == team, 'teamDefRating'] = teamDef[team]

def relRatings(dataFrame):
    dataFrame['relToLeagueOffRating'] = (dataFrame['OnOffRtg'] - dataFrame['avgRating']) / dataFrame['OnOffRtg'].std()
    dataFrame['relToLeagueDefRating'] = (dataFrame['OnDefRtg'] - dataFrame['avgRating']) / dataFrame['OnDefRtg'].std()
    dataFrame['relToTeamOffRating'] = (dataFrame['OnOffRtg'] - dataFrame['teamOffRating']) / dataFrame['OnOffRtg'].std()
    dataFrame['relToTeamDefRating'] = (dataFrame['OnDefRtg'] - dataFrame['teamDefRating']) / dataFrame['OnDefRtg'].std()
```

```python
def averageOfColByYear(dataFrame, column):
    newColumn = 'relToAvg_'
    newColumn += column
    temporary1 = dataFrame.groupby('year')[column].mean()
    temporary2 = dataFrame.groupby('year')[column].std()
    for year in temporary1.index:
        dataFrame.loc[dataFrame['year'] == year, newColumn] = (dataFrame.loc[dataFrame['year'] == year, column] - temporary1[year]) / temporary2[year]

def averageOfColByYearByTeam(dataFrame, column):
    newColumn = 'relToTeamAvg_'
    newColumn += column
    temporary1 = dataFrame.groupby(['year','team'])[column].mean()
    temporary2 = dataFrame.groupby(['year','team'])[column].std()
    for tuple in temporary1.index:
        dataFrame.loc[(dataFrame['year'] == tuple[0]) & (dataFrame['team'] == tuple[1]), newColumn] = (dataFrame.loc[(dataFrame['year'] == tuple[0]) & (dataFrame['team'] == tuple[1]), column] - temporary1[tuple]) / temporary2[tuple]
    
```

```python
def regressionModel(dataFrame, dependentVariable, independentVariable):
    Formula = 'dep ~ ind'
    Formula = Formula.replace('dep', dependentVariable)
    Formula = Formula.replace('ind', independentVariable)

    model = smf.ols(formula = Formula, data = dataFrame)
    res = model.fit()
    return res
```

```python
possIndVars = ['fieldGoals', 'fieldGoalAttempts', 'fieldGoalPercentage', 'threePointers', 'threePointAttempts',
               'threePointPercentage', 'twoPointers', 'twoPointAttempts','twoPointPercentage', 'freeThrows', 
               'freeThrowAttempts', 'freeThrowPercentage', 'offensiveReb', 'totalReb', 'assists', 'steals',
               'blocks', 'turnovers', 'personalFouls', 'points', 'pointsPerScoringAttempt',
               'relToAvg_fieldGoals', 'relToAvg_fieldGoalAttempts', 'relToAvg_fieldGoalPercentage', 'relToAvg_threePointers', 
               'relToAvg_threePointAttempts', 'relToAvg_threePointPercentage', 'relToAvg_twoPointers', 'relToAvg_twoPointAttempts',
               'relToAvg_twoPointPercentage', 'relToAvg_freeThrows', 'relToAvg_freeThrowAttempts', 'relToAvg_freeThrowPercentage', 
               'relToAvg_offensiveReb', 'relToAvg_totalReb', 'relToAvg_assists', 'relToAvg_steals',
               'relToAvg_blocks', 'relToAvg_turnovers', 'relToAvg_personalFouls', 'relToAvg_points', 'relToAvg_pointsPerScoringAttempt',
               'relToTeamAvg_fieldGoals', 'relToTeamAvg_fieldGoalAttempts', 'relToTeamAvg_fieldGoalPercentage', 
               'relToTeamAvg_threePointers', 'relToTeamAvg_threePointAttempts', 'relToTeamAvg_threePointPercentage', 
               'relToTeamAvg_twoPointers', 'relToTeamAvg_twoPointAttempts','relToTeamAvg_twoPointPercentage', 'relToTeamAvg_freeThrows', 
               'relToTeamAvg_freeThrowAttempts', 'relToTeamAvg_freeThrowPercentage', 'relToTeamAvg_offensiveReb', 
               'relToTeamAvg_totalReb', 'relToTeamAvg_assists', 'relToTeamAvg_steals', 'relToTeamAvg_blocks', 
               'relToTeamAvg_turnovers', 'relToTeamAvg_personalFouls', 'relToTeamAvg_points', 'relToTeamAvg_pointsPerScoringAttempt',]

possDepVars = ['PlusMinus', 'OnOffRtg','OnDefRtg','teamOffRating', 'teamDefRating', 'relToLeagueOffRating', 
              'relToLeagueDefRating', 'relToTeamOffRating', 'relToTeamDefRating', 'teamVSleagueOff']

guardReg = []
for i in range(len(possDepVars)):
    for j in range(len(possIndVars)):
        guardReg.append(regressionModel(guards, possDepVars[i], possIndVars[j]))

forwardReg = []
for i in range(len(possDepVars)):
    for j in range(len(possIndVars)):
        forwardReg.append(regressionModel(forwards, possDepVars[i], possIndVars[j]))

centerReg = []
for i in range(len(possDepVars)):
    for j in range(len(possIndVars)):
        centerReg.append(regressionModel(centers, possDepVars[i], possIndVars[j]))
```

```python
def highRSquared(list):
    out = []
    for i in range(len(list)):
        if list[i].rsquared >= 0.1:
            out.append(list[i])
    return out

def maxRSquared(list):
    out = 0
    for i in range(len(list)):
        if list[i].rsquared >= 0:
            out = list[i].rsquared
    return int(out)
```

```python
guardHighRSquared = highRSquared(guardReg)
forwardHighRSquared = highRSquared(forwardReg)
centerHighRSquared = highRSquared(centerReg)
```
