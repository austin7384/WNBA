```python
import numpy as np
import pandas as pd
from scipy.stats import boxcox
from sklearn.preprocessing import scale
```

```python
playersBox2024 = pd.read_csv('WNBA Stats/playersBox2024.csv')
playersPBP2024 = pd.read_csv('WNBA Stats/2024.csv')
```

```python
del playersBox2024['Unnamed: 0']
playersBox2024.head()
```

```python
playersPBP2024.head()
```

```python
playersPBP2024 = playersPBP2024.iloc[:, :7]
playersPBP2024['PlusMinus'] = playersPBP2024['OnOffRtg'] - playersPBP2024['OnDefRtg']
playersPBP2024.head()
```

```python
players2024 = pd.merge(playersBox2024, playersPBP2024, left_on = 'player', right_on = 'Name', how = 'left')
players2024.head()
```

```python
playersFeature2024 = players2024.drop(['player', 'team', 'position', 'gamesPlayed', 'minutesPlayed', 'gamesStarted', 'year', 'Name', 'TeamAbbreviation', 'GamesPlayed', 'Minutes'], axis = 1)
playersFeature2024.head()
```

```python
#Select columns with only positive elements
pos_indx = playersFeature2024.apply(lambda x: np.all(x>0))


# Reduce skewness of positive columns using Box-Cox and add back to features dataframe
playersFeature2024.loc[:, pos_indx] = playersFeature2024.loc[:, pos_indx].apply(lambda x: boxcox(x)[0])
playersFeature2024.head()
```
