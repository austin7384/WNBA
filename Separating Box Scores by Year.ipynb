```python
import pandas as pd
import numpy as np
```

```python
players = pd.read_csv('WNBA Box Score Stats 2009-2024 - Sheet1.csv')
players = players[(players['minutesPlayed'] >= 400) & (players['team'] != 'TOT')]
players.head()
```

```python
players.shape
```

```python
# get points per shot attempt
players['shotAttempts'] = np.round(players['fieldGoalAttempts'] + (0.44 * players['freeThrowAttempts']), 1)
players['pointsPerShot'] = np.round(players['points'] / players['shotAttempts'], 5)
players.head()
```

```python
for year in range(2009, 2025, 1):
    csvName = 'playersBox' + str(year) + '.csv'
    temp = players[players['year'] == year]
    temp.to_csv(csvName)
```
